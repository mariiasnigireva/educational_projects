Разделим данные на две группы:  тех, кто ушел в отток и тех, кто остался. Посмотрим, как изменились средние значения в двух группах:

Мы видим, что такие признаки, как пол и наличие контактного телефона не отличаются в двух группах, значит, их можно не учитывать в дальнейшем исследовании. Среднее значение для признака Near_Location меняется всего на 10%, поэтому тоже является претендентом на выбывание. Посмотрим, как меняются распределения остальных признаков в двух группах:

```python
def plot_distributions(dataframes, labels, columns, height=3):
    """
    Функция берёт массив датасетов, их названия для легенды и массив столбцов по которым нужно построить гистограмму.
    Далее функция создает поле для графиков и строит столбчатые гистограммы распределения данных датасетов для каждого столбца.
    Количество используемых бинов гистограммы не более 12.
    Возвращает построенный график.
    """
    num_rows = int(np.ceil(len(columns)/3))
    fig, axs = plt.subplots(num_rows,3, figsize=(15,height*num_rows))
    if num_rows == 1:
        axs = [axs]

    for i, col in enumerate(columns):
        vmin = gym_churn[col].min()
        vmax = gym_churn[col].max()
        step = 1
        if (vmax-vmin)/step > 12:
            step = (vmax-vmin)/12
        ax = axs[i//3][i%3]
        ax.hist([df[col] for df in dataframes], bins=np.arange(vmin,vmax+1.01,step), density=True, label=labels, align='left')
        ax.set_title(f'Параметр {col}')
        ax.set_ylabel('Доля клиентов')
        ax.set_xlabel('Значение параметра')
        ax.legend()
        for axis in [ax.xaxis, ax.yaxis]:
            axis.set_major_locator(ticker.MaxNLocator(integer=True))
    plt.tight_layout()
    return fig
```

```python
# делим данные на две части - покинувшие культуристов и оставшиеся
gym_left = gym_churn.loc[gym_churn['Churn']==1]
gym_active = gym_churn.loc[gym_churn['Churn']==0]
# столбцы для графика
columns = [
    'Near_Location', 'Partner', 'Promo_friends', 'Phone',
    'Contract_period', 'Group_visits', 'Age',
    'Avg_additional_charges_total', 'Month_to_end_contract', 'Lifetime',
    'Avg_class_frequency_total', 'Avg_class_frequency_current_month',
]
# строим распределение
fig = plot_distributions([gym_left, gym_active], ['покинули', 'остались'], columns)
fig.suptitle('Распределение параметров для клиентов, покинувших фитнес-студию и оставшихся:', fontsize=16, y=1.02)
plt.show()
```

По построенным гистограммам распределений признаков видим, что некоторые признаки распределены для обоих групп клиентов примерно одинаково: это признаки расположения "Near_Location" и 'Promo_friends', их также не будем учитывать при дальнейшем анализе. 

Заметно, что распределения на двух последних графиках почти одинаковы. Это означает, что мы можем использовать вместо обоих признаков либо только один, либо разницу между ними. Если посмотреть на таблицу gym_churn_mean, видно, что среди ушедших пользователей упало среднее число посещений за последний месяц. Создадим новый признак Avg_class_frequency_dif, вычтя из Avg_class_frequency_total значения Avg_class_frequency_current_month, он будет означать, на сколько в среднем изменилось посещение студии за последний месяц по сравнению с обычным показателем:

```python
gym_churn['Avg_class_frequency_dif'] = gym_churn['Avg_class_frequency_total'] - gym_churn['Avg_class_frequency_current_month']
```

* 'Month_to_end_contract' - сколько месяцев осталось до конца контракта, 
* 'Lifetime' - как долго человек является клиентом студии, 
* 'Avg_class_frequency_dif' - на сколько отличается посещение студии клиентом в среднем и за последний месяц. 

## Шаг 3. Построение модели прогнозирования оттока пользователей  <a id="step9"></a>

Поскольку нам изначально известно, останется ли клиент на следующий месяц в фитнес-студии, а варианта всего два, то необходимо построить модель бинарной классификации пользователей (целевой признак — факт оттока пользователя в следующем месяце), но сначала разделим датасет на обучающую и валидационную выборку в стандартных пропорциях 80 к 20:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Для бинарной классификации в обучении с учителем применим модели логистической регрессии и случайный лес. Это достаточно простые модели (не требующие слишком больших технических и временных ресурсов), отличающиеся хорошим предсказательным уровнем. Обучим модель на train-выборке обоими способами. Для сравнения моделей будем использовать стандартные метрики accuracy - доля правильных ответов, precision - точность и recall - полнота:

```python
models = [LogisticRegression(random_state=0), RandomForestClassifier(random_state=0)]

# функция, которая принимает на вход модель и данные и выводит метрики
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Accuracy:{:.2f} Precision:{:.2f} Recall:{:.2f} '.format(accuracy_score(y_test, y_pred), 
                                precision_score(y_test, y_pred),
                                recall_score(y_test, y_pred)))
    return model
```

```python
ms = []
# цикл, который выводит метрики по списку моделей
for i in models:
    print(i)
    m = make_prediction(i, X_train, y_train, X_test, y_test)
    ms.append(m)
```

```python
pd.DataFrame([X.columns.to_list(),  ms[1].feature_importances_]).T\
    .rename(columns={0: 'Признак', 1: 'Важность'})\
    .sort_values(by='Важность', ascending=False)\
    .style\
    .bar(['Важность'])
```

Для нашей бизнес-задачи наиболее важна метрика precision, поскольку мы будем вкладывать больше ресурсов отдела маркетинга в клиентов, целевой признак для которых равен единице. Но и доля верно угаданных ответов из всех прогнозов (accuracy) и precision лучше для модели RandomForestClassifier. Кроме того, сама по себе доля верно угаданных ответов очень высока.

## Шаг 4. Кластеризация пользователей  <a id="step11"></a> 

К настоящему моменту с помощью модели случайный лес мы умеем предсказывать, кто из наших клиентов уйдёт, значит мы можем применить к ним меры по удержанию. Но клиенты фитнес-студии неоднородны: у них разный пол, возраст, они с разной частотой посещают групповые занятия и саму студию. Для того, чтобы применять методы маркетинга более точечно стоит сегментировать клиентов. Для проведения кластеризации нужно сначала стандартизировать данные (привести их к единому масштабу), а затем построить и визуализировать матрицу расстояний: 

```python
# создадим матрицу всех признаков
X_all = gym_churn.drop(columns = ['Churn'])
# создадим объект класса StandardScaler и применим его к 
scaler = StandardScaler()
X_all_st = scaler.fit_transform(X_all)
```

Теперь можно построить матрицу расстояний и нарисовать дендрограмму:

```python
linked = linkage(X_all_st, method = 'ward')
```

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
ax.set_title("Дендрограмма кластеризации пользователей сети фитнес-студий")
plt.show() 
```

Судя по цветам в отрисованной дендрограмме, рекомендуемое число кластеров - не менее четырёх (обозначены разными цветами на графике). Но один из кластеров слишком большой, поэтому чтобы улучшить точность наших маркетинговых стратегий, поделим наших клиентов на **пять кластеров**. На основании алгоритма K-Means обучим модель кластеризациии и спрогнозируем кластеры клиентов:

```python
km = KMeans(n_clusters = 5) # задаём число кластеров, равное 5
labels = km.fit_predict(X_all_st) # применяем алгоритм к данным и формируем вектор кластеров 
# сохраняем метки кластера в поле датасета
gym_churn['cluster_km'] = labels
# выводим статистику по средним значениям признаков по кластеру
gym_churn.groupby(['cluster_km']).mean()
```

Судя по средним значениям признаков по кластерам можно сделать вывод, что в каждом кластере примерно поровну мужчин и женщин. Также очень слабо меняются по кластерам средние значения признаков Near_Location, Age и Avg_additional_charges_total. А вот с наличием номера телефона ситуация другая: в одном из кластеров никто из клиентов не оставил свой номер, тогда как в остальных четырёх почти все оставляли контакт. Также кардинальные различия по кластерам видны для признаков Contract_period, Month_to_end_contract и Lifetime. 

Посмотрим теперь на распределения тех признаков, которые сильно различаются по кластерам, речь идёт о распределении внутри кластера:

```python
important_features_clusters = ['Group_visits', 'Month_to_end_contract', 'Lifetime']

for k in range (0, 5):
    df = gym_churn.loc[gym_churn['cluster_km'] == k] # делим данные на кластеры
    print(f'Распределения параметров для кластера номер {k}')
    plot_distributions([df], ['распределение параметра'], important_features_clusters) # строим распределение
    plt.show()
```

Видно, что каждый из параметров наиболее отличен в одном из кластеров. Хорошо бы понять, из какого кластера уходит больше всего клиентов и посмотреть на распределения признаков в нём. Но сначала построим также распределения для 'Age', 'Avg_additional_charges_total', 'Avg_class_frequency_dif' (даже при одинаковых средних значениях распределения могут сильно отличаться):

```python
important_features_clusters_2 = ['Age', 'Avg_additional_charges_total', 'Avg_class_frequency_dif']
for k in range (0, 5):
    df = gym_churn.loc[gym_churn['cluster_km'] == k] # делим данные на кластеры
    print(f'Распределения параметров для кластера номер {k}')
    plot_distributions([df], ['распределение параметра'], important_features_clusters_2) # строим распределение
    plt.show()
```

Итак, зная, из каких кластеров уходит больше всего клиентов можно по графикам распределений выделить ключевые отличия этого кластера. Посмотрим теперь из каких кластеров уходит больше всего клиентов. В построенной выше таблице в столбце Churn подсчитана доля оттока клиентов из кластеров. Выделим её в отдельную переменную и посмотрим ещё раз на результат:

```python
clusters_churn = gym_churn.groupby(['cluster_km']).mean().Churn
clusters_churn
```

Мы видим сильные различия в доле оттока по кластерам: два кластера довольно надёжны, поскольку из них уйдёт от трёх до восьми процентов клиентов (на них точно можно сэкономить ресурсы отдела маркетинга), ещё два кластера более склонны к оттоку, а вот клиенты из последнего кластера точно покинут фитнес-студию, на них нужно расходовать основную часть бюджета. Посмотрим, сколько клиентов из нашей выборки в каждом кластере:

```python
gym_churn.cluster_km.value_counts()
```

Мы видим, что число людей в кластерах сильно разнится. Узнаем, какова доля клиентов, которые уйдут (от всех ушедших из фитнес-студии) сосредоточена в кластере с максимальным оттоком. Сначала выведим его номер (при повторном запуске модели номера кластеров меняются):

```python
n_churn = clusters_churn.argmax()
print(f'Из кластера номер {n_churn} уйдут почти все клиенты.')
```

Подсчитаем, какая доля клиентов, которые уйдут (от всех ушедших) сосредоточена в этом кластере: 

```python
gym_churn[gym_churn['cluster_km'] == n_churn].Churn.value_counts()[1] # сколько клиентов и нужного класса уйдут
gym_churn.Churn.value_counts()[1] # сколько клиентов уйдёт всего
ratio = gym_churn[gym_churn['cluster_km'] == n_churn].Churn.value_counts()[1] / gym_churn.Churn.value_counts()[1]
print(f'Из рассматриваемого класса уйдёт {round(ratio * 100, 1)}% всех клиентов.')
```
