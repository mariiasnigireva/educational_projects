–ü–æ–ª—É—á–∏–ª–∏ –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç, —á—Ç–æ —Å–∞–º–æ–µ –±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º –≤ —Å—Ç–æ–ª–æ–≤—ã—Ö. –î–∞–ª–µ–µ —Å–ª–µ–¥—É—é—Ç —Ä–µ—Å—Ç–æ—Ä–∞–Ω—ã, –±—É—Ñ–µ—Ç—ã –∏ –±–∞—Ä—ã. –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ –∫–∞—Ñ–µ —Ä–∞–≤–Ω–æ —Å–æ—Ä–æ–∫–∞. –ü–æ—Å–∫–æ–ª—å–∫—É –º—ã —Ö–æ—Ç–∏–º –æ—Ç–∫—Ä—ã—Ç—å –Ω–µ–±–æ–ª—å—à–æ–µ –∫–∞—Ñ–µ, –±—É–¥–µ–º –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å—Å—è –Ω–∞ –æ–ø—ã—Ç –∫–æ–Ω–∫—É—Ä–µ–Ω—Ç–æ–≤, —Ç.–µ. **—Ä–µ–∫–æ–º–µ–Ω–¥—É–µ–º—ã–π —Ä–∞–∑–º–µ—Ä –∑–∞–≤–µ–¥–µ–Ω–∏—è - –¥–æ 40 –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç**.

–ò—Ç–∞–∫, –º—ã –æ–ø—Ä–µ–¥–µ–ª–∏–ª–∏—Å—å —Å —Ç–∏–ø–æ–º –∏ —Ä–∞–∑–º–µ—Ä–æ–º –∑–∞–≤–µ–¥–µ–Ω–∏—è, —Ç–µ–ø–µ—Ä—å –∏–∑—É—á–∏–º –ø–æ –ø–æ–¥—Ä–æ–±–Ω–µ–µ —Ç–µ—Ä—Ä–∏—Ç–æ—Ä–∏–∞–ª—å–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫, —á—Ç–æ–±—ã –≤—ã–±—Ä–∞—Ç—å –ø–æ–¥—Ö–æ–¥—è—â–µ–µ –º–µ—Å—Ç–æ. –£–∑–Ω–∞–µ–º, –∫–∞–∫–∏–µ —É–ª–∏—Ü—ã –∏ —Ä–∞–π–æ–Ω—ã —Å–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ. –í—ã–¥–µ–ª–∏–º –¥–ª—è –Ω–∞—á–∞–ª–∞ –≤ –Ω–∞—à–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ —É–ª–∏—Ü—É –∏–∑ –ø–æ–ª—è —Å –∞–¥—Ä–µ—Å–æ–º:

```python
def get_street(address):
    """
    –§—É–Ω–∫—Ü–∏—è –±–µ—Ä—ë—Ç –ø–æ–ª–µ —Å –∞–¥—Ä–µ—Å–æ–º, —Ä–∞–∑–¥–µ–ª—è–µ—Ç –µ–≥–æ –ø–æ –∑–∞–ø—è—Ç–æ–π, –≤ –∫–∞—á–µ—Å—Ç–≤–µ —É–ª–∏—Ü—ã –ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è
    –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –∑–∞–ø—è—Ç–æ–π, –ø–æ—Å–∫–æ–ª—å–∫—É, –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ, –≤ –∞–¥—Ä–µ—Å–µ —É–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å–Ω–∞—á–∞–ª–∞ –≥–æ—Ä–æ–¥, –∑–∞—Ç–µ–º —É–ª–∏—Ü–∞.
    –ï—Å–ª–∏ –≤ –∞–¥—Ä–µ—Å–µ –Ω–µ —Ñ–∏–≥—É—Ä–∏—Ä—É–µ—Ç —Å–ª–æ–≤–æ "–≥–æ—Ä–æ–¥", —Ç–æ —ç—Ç–æ –∞–¥—Ä–µ—Å —Å–µ–ª–∞, –ø–æ—Å—ë–ª–∫–∞, –∏–ª–∏ –¥—Ä—É–≥–æ–≥–æ –Ω–µ–±–æ–ª—å—à–æ–≥–æ
    –ø–æ—Å–µ–ª–µ–Ω–∏—è, –≥–¥–µ –Ω–µ —Ç–∞–∫ –º–Ω–æ–≥–æ —É–ª–∏—Ü. –í —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ, –±–µ—Ä—ë–º –≤ –∫–∞—á–µ—Å—Ç–≤–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —É–ª–∏—Ü—ã - –≥–ª–∞–≤–Ω–∞—è —É–ª–∏—Ü–∞.
    """
    address_list = address.split(',')
    if '–ú–æ—Å–∫–≤–∞' not in address_list[0]:
        street = '–Ω–µ –ú–æ—Å–∫–≤–∞'
    elif '–≥–æ—Ä–æ–¥' not in address_list[0]:
        street = '–≥–ª–∞–≤–Ω–∞—è —É–ª–∏—Ü–∞'
    else:
        street = address_list[1].strip()
    return street
```

```python
rest_data_unique['street'] = rest_data_unique['address'].apply(get_street)
```
