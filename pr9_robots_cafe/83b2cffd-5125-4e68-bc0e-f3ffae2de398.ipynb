Поскольку мы хотим открыть небольшое кафе, будем ориентироваться на опыт конкурентов, т.е. **рекомендуемый размер заведения - до 40 посадочных мест**.

Итак, мы определились с типом и размером заведения, теперь изучим по подробнее территориальный признак, чтобы выбрать подходящее место. Узнаем, какие улицы и районы самые популярные. Выделим для начала в нашем датафрейме улицу из поля с адресом:

```python
def get_street(address):
    """
    Функция берёт поле с адресом, разделяет его по запятой, в качестве улицы принимается информация
    после первой запятой, поскольку, как правило, в адресе указывается сначала город, затем улица.
    Если в адресе не фигурирует слово "город", то это адрес села, посёлка, или другого небольшого
    поселения, где не так много улиц. В таком случае, берём в качестве название улицы - главная улица.
    """
    address_list = address.split(',')
    if 'Москва' not in address_list[0]:
        street = 'не Москва'
    elif 'город' not in address_list[0]:
        street = 'главная улица'
    else:
        street = address_list[1].strip()
    return street
```

```python
rest_data_unique['street'] = rest_data_unique['address'].apply(get_street)
```
